```python
## --------- TASK --------------
##Write a Python function to calculate the covariance matrix for a given set of vectors. 
# The function should take a list of lists, where each inner list represents a feature 
# with its observations, and return a covariance matrix as a list of lists.

def covariance(a: list[float], b: list[float]):
	##Calculating mean for each list/variable:
	mean_a = 0
	for i in a:
		mean_a+= i
	mean_a = mean_a/len(a)

	mean_b = 0
	for i in b:
		mean_b+= i
	mean_b = mean_b/len(b)

	prod = 0
	##formula for sample covariance sum[(a_i - mean_a) * (b_i - mean_b)] / N-1
	for i in range(len(a)):
		prod += ((a[i]-mean_a)* (b[i]-mean_b))
	E_AB = prod/(len(a)-1)

	cov_ab = E_AB 
	return cov_ab

	


def calculate_covariance_matrix(vectors: list[list[float]]) -> list[list[float]]:
	##making the covariance matrix
	cov_m = []
	for i in range(len(vectors)):
		cov_m.append([])
	
	##Inserting values of covariance
	for i in range(len(vectors)):
		for j in range(len(vectors)):
			cov_m[i].append(covariance(vectors[i], vectors[j]))

	return cov_m
```

```python
print(calculate_covariance_matrix([[1, 2, 3], [4, 5, 6]]))
```
