```python
"""
GOAL:
--------------------------------------------------------------------------------
# Return a list where each element is the dot product of a row of 'a' with 'b'.
# If the number of columns in 'a' does not match the length of 'b', return -1.
-------------------------------------------------------------------------------
"""

def matrix_dot_vector(a: list[list[int|float]], b: list[int|float]) -> list[int|float]:
       
    ##Check if number of columns in matrix "a" are at the same length as the vector "b". Here we try make sure that this is true for evert row.
       for i in range(len(a)):
            if len(a[i]) != len(b):
                return -1
        
    ##First we make a new vector that holds the dot product(nx1). This will have length equal to rows in matrix "a"
       new_vec = []
       for i in range(len(a)):
            new_vec.append(0)
    
    ##The dot product calculation 
       for i in range(len(b)):
           for p in range(len(a)):
               new_vec[p] += (b[i] * a[p][i])
			
       return new_vec
```

```python
#This should print: [14, 25, 49]
print(matrix_dot_vector([[1, 2, 3], [2, 4, 5], [6, 8, 9]], [1, 2, 3]))

##This should print: [14, 49]
print(matrix_dot_vector([[1, 2, 3], [6, 8, 9]], [1, 2, 3]))
```

```python
#This should print: -1 since the second row only has two columns
print(matrix_dot_vector([[1, 1, 1], [2, 2], [3, 3, 3]], [2, 2, 2]))
```
