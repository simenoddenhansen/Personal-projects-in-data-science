### Dot product of two 1d numpy arrays <br>
Args: <br>
    vec1 (numpy.ndarray): 1d array representing the first vector <br>
    vec2 (numpy.ndarray): 1d array representing the second vector <br><br>
Returns:<br>
    The dot product of the two vectors.

```python
import numpy as np

def calculate_dot_product(vec1, vec2):
    ##Calculating the dot products. 
	dot_prod = sum([j*i for j,i in zip(vec1,vec2)])
	return dot_prod
```

```python
##TEST: output should be 32
vec1 = [1, 2, 3]
vec2 = [4, 5, 6]
print(calculate_dot_product(vec1, vec2))
```
