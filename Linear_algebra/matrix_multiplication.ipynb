The goal is to make a function that multiplies two matrices together and returns -1 if shapes of matrix don't align.

```python
#This takes two matrices a and b and return the product of the two matrices. 
#The requirement for this operation to be valid is that matrix a has as many columns
# as b has rows. 

def matrixmul(a:list[list[int|float]], b:list[list[int|float]])-> list[list[int|float]]:
    ##Check if n_cols_a = n_rows_b, otherwise return -1. Here check if all rows of a has the same 
    # amount of values(column values) as b has rows. 
    for i in a:
        if len(i) != len(b):
            return -1
    
    ##Making the new matrix according to the correct dimensions nxm where n=#(rows of a) and m = #(cols of b)
    rows_a = len(a)
    cols_b = len(b[0])
    c = [[0 for i in range(cols_b)] for i in range(rows_a)]

    #Iterate over rows of a
    for i in range(len(a)):
    #Iterate over the columns of b
        for j in range(len(b[0])):
            #Iterate over rows of b
            for t in range(len(b)):
                c[i][j] += a[i][t] * b[t][j]
    return c
```
