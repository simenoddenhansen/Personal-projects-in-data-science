```python
def transpose_matrix(a: list[list[int|float]]) -> list[list[int|float]]:
    new_matrix = []

    ##This adds number of rows as the number of columns from the old matrix
    for i in range(len(a[0])):
        new_matrix.append([])

    ##This take every element in each column in "a" and adds it to the correct row in the new matrix
    for i in range(len(a)): #looping through the rows in a
        for j in range(len(a[i])): 
            #For every element in each row of "a", we now append them to the j'th row-vector/list in "a-transpose".
            new_matrix[j].append(a[i][j])     
    return new_matrix
```

```python
## We expect this to write: [[1, 3, 5], [2, 4, 6]]
print(transpose_matrix([[1, 2], [3, 4], [5, 6]]))
```
