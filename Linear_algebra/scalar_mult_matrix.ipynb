The goal is to make a python function that multiplies a matrix by a scalar

```python
def scalar_multiply(matrix: list[list[int|float]], scalar: int|float) -> list[list[int|float]]:
	result = []
	
	##Adds the rows to the resulting matrix
	for row in matrix:
		result.append(row.copy())
	
	##Multiplies each element in "result" with the scalar
	for i in range(len(result)):
		for j in range(len(result[i])):
			result[i][j] = result[i][j]*scalar
			j += 1
		
	return result
```

```python
matrix = [[1/2,1],[1/3,1]]
scalar_multiply(matrix, 4)
```
